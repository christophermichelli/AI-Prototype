```python
def snowboarding_conditions(snowfall_inches, temperature_f, wind_mph):
    score = 0

    # Snowfall evaluation
    if snowfall_inches >= 6:
        score += 3
    elif snowfall_inches >= 3:
        score += 2
    elif snowfall_inches >= 1:
        score += 1

    # Temperature evaluation
    if 20 <= temperature_f <= 30:
        score += 3
    elif 10 <= temperature_f < 20 or 30 < temperature_f <= 35:
        score += 2
    else:
        score += 0

    # Wind evaluation
    if wind_mph <= 10:
        score += 3
    elif wind_mph <= 20:
        score += 1
    else:
        score -= 1

    # Final decision
    if score >= 7:
        return "ğŸ‚ Excellent conditions! Fresh snow and great weather."
    elif score >= 4:
        return "ğŸ™‚ Good conditions. Worth hitting the slopes."
    else:
        return "âš ï¸ Poor conditions. Might want to wait for better weather."


# Example usage
snowfall = float(input("Snowfall (inches): "))
temperature = float(input("Temperature (Â°F): "))
wind = float(input("Wind speed (mph): "))

result = snowboarding_conditions(snowfall, temperature, wind)
print("\nSnowboarding Report:")
print(result)
```
